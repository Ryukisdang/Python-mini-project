```python
def start_game():
    print("You wake up in a dimly lit room. It feels cold and eerie.")
    print("You realize that you are trapped in a haunted house.")
    print("Your objective is to find a way to escape.")
    print("Good luck!\n")

    room = "kitchen"
    escape = False

    while not escape:
        if room == "kitchen":
            print("You are in the kitchen.")
            print("There are two doors - one on the left and one on the right.")
            choice = input("Which door do you choose? (left/right) ")

            if choice == "left":
                room = "living room"
            elif choice == "right":
                room = "dining room"
            else:
                print("Invalid input. Try again.")

        elif room == "living room":
            print("You are in the living room.")
            print("There is a spooky painting on the wall and a locked door ahead.")
            choice = input("What do you do? (inspect/picklock) ")

            if choice == "inspect":
                print("As you closely examine the painting, you notice a hidden key.")
                print("You take the key and proceed to the next room.")
                room = "hallway"
            elif choice == "picklock":
                print("You attempt to picklock the door, but fail.")
                print("A trap door opens beneath you, and you fall into a bottomless pit.")
                print("Game Over.")
                return

        elif room == "dining room":
            print("You are in the dining room.")
            print("There is a table set with a feast and a staircase leading upstairs.")
            choice = input("What do you do? (eat/take stairs) ")

            if choice == "eat":
                print("You sit down to eat, but the food turns into worms.")
                print("You vomit uncontrollably and pass out.")
                print("Game Over.")
                return
            elif choice == "take stairs":
                room = "bedroom"
        elif room=='bedroom':
            print("You are in the bedroom")
            print("There is a wardrobe and a hallway.")
            choice = input("What do you do? (wardrobe/under the bed) ")
            if choice=='wardrobe':
                print("There was a bear trap")
                print("You died from extreme pain and heavy loss of blood")
                print("Game over")
                break
            if choice=='under the bed':
                print("You found and old brief case")
                print('what do you want to do with it?')
                choice=input("open it or leave it: ")
                if choice=='open it':
                    print("you found a photograph of the person who is deceased and a serial killer also the owner of the house...")
                    print("The past story of this house about the serial killer owner will be covered..To be continued................\nHahahahahahaah")
                    break
                if choice=='leave it':
                    print("You just a heard someone walking step sound above the floor of the bedroom")
                    print("You just a goosebumps, do you have to courage go upstair yes or no?")
                    choice=input("yes or no>> ")
                    if choice=='yes':
                        room='hallway'
                    if choice=='no':
                        print("There was ghost of serial killer behind you that twist your neck\nyou died disgusting,.EMOTIONALLLLL DAMAGEEEE!!!!")
                        break
                    
                
                

        elif room == "hallway":
            print("You are in a dimly lit hallway.")
            print("There are three doors - red, blue, and green.")
            choice = input("Which door do you choose? (red/blue/green) ")

            if choice == "red":
                print("As you open the red door, a ghost appears and scares you to death.")
                print("Game Over.")
                return
            elif choice == "blue":
                room = "bathroom"
            elif choice == "green":
                room = "backyard"
            else:
                print("Invalid input. Try again.")

        elif room == "bathroom":
            print("You are in the bathroom.")
            print("There is a mirror on the wall and a window with rusty bars.")
            choice = input("What do you do? (look in mirror/escape through window) ")

            if choice == "look in mirror":
                print("Your reflection in the mirror starts screaming.")
                print("You lose your sanity and go insane.")
                print("Game Over.")
                return
            elif choice == "escape through window":
                print("You manage to squeeze through the rusty bars and climb outside.")
                room = "backyard"

        elif room == "backyard":
            print("You are in the backyard")
```
